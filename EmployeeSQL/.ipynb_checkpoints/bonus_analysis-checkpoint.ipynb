```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
from config import username, password
```

```python
#1. Import the SQL database into Pandas. 
# * Consult [SQLAlchemy documentation](https://docs.sqlalchemy.org/en/latest/core/engines.html#postgresql) for more information.
# * If using a password, do not upload your password to your GitHub repository. See [https://www.youtube.com/watch?v=2uaTPmNvH0I](https://www.youtube.com/watch?v=2uaTPmNvH0I) and [https://help.github.com/en/github/using-git/ignoring-files](https://help.github.com/en/github/using-git/ignoring-files) for more information.

engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/Employee')
connection = engine.connect()
```

```python
#importing employee table for analysis
employee = pd.read_sql('SELECT * FROM employee', connection)
employee
```

```python
#importing salaries table for analysis
salaries = pd.read_sql('SELECT * FROM salaries', connection)
salaries
```

```python
#importing title table for analysis
title = pd.read_sql('SELECT * FROM title', connection)
title
```

```python
#merging employee and salaries table on emp num for analysis
emp_sal = employee.merge(salaries, on = 'emp_no')
emp_sal
```

```python
#merging table created above with title table for analysis
emp_sal_title = emp_sal.merge(title, left_on = 'emp_title', right_on = 'title_id')
emp_sal_title
```

```python
#creating table with desired coloumns
salary_df = emp_sal_title[["salary", "title"]]
salary_df
```

```python
#2. Create a histogram to visualize the most common salary ranges for employees.
plt.figure(figsize=(10,5))
bins = [30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000]
plt.hist(x=salary_df["salary"], bins = bins, color = 'b', alpha=0.9, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Salary Range')
plt.ylabel('Frequency')
plt.title('Most Common Salary Ranges')
plt.show()
```
